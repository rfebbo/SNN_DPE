```python
import matplotlib.pyplot as plt
import numpy as np

import json
from matplotlib.lines import Line2D

import os
```

```python
folder = './runs/1000_runs_10_connectivities/'

input_distances = []
output_distances = []
connectivities = []
for file in os.listdir(folder):
    with open(folder + file, 'r') as f:
        run = {}
        for l in f:
            run = json.loads(l)
        
        try:
            connectivities.append(run['connectivity'])
            input_distances.append(run['input distances'])
            output_distances.append(run['output distances'])
        except:
            print(f'file: {file} didnt work')


```

```python
for c_i in range(len(connectivities)):
    OD = {}
    for run in range(len(input_distances[c_i])):
        for id, od in zip(input_distances[c_i][run], output_distances[c_i][run]):
            if id not in OD:
                OD[id] = []

            OD[id].append(od)

    OD_l = []
    for id in OD:
        OD_l.append(OD[id])

    slc = slice(0,None,3)
    plt.violinplot(OD_l[slc], positions=list(OD.keys())[slc], widths=0.01)
    plt.title(f'{c_i}')
    plt.show()
    # plt.scatter(input_distances[0], output_distances[0])
```

```python

colors = ['tab:blue', 'tab:orange', 'tab:green', 'tab:red', 'tab:purple']
legend_elements = []
con_out_ds = []
fig, ax = plt.subplots(figsize=(10,10))

for d, c, col in zip(distances[0:-1:10], connectivities[0:-1:10], colors):
    legend_elements.append(Line2D([0], [0], marker='o', color=col, label=c, markerfacecolor=col, markersize=5))
    out_ds = []
    for input_d in d:
        out_ds.append(len(np.unique(d[input_d])))
        plt.scatter([input_d] * len(d[input_d]), d[input_d], label=c, c=col)

    con_out_ds.append(out_ds)


plt.legend(handles=legend_elements)
# fig, ax = plt.subplots(len(con_out_ds))
# for con, axi in zip(con_out_ds, ax):
    # axi.scatter(range(len(con)), con)
```
