```python
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams.update({'figure.figsize': (5,3)})

from HD_main import run_distance_test

p_con = 0.03
run = run_distance_test(connectivity=p_con, n_enc=5, n_neurons=20, n_runs=500, save_folder=None, n_proc=10)
```

```python

OD = {}
for run_n in range(len(run['input distances'])):
    for id, od in zip(run['input distances'][run_n], run['output distances'][run_n]):
        if id not in OD:
            OD[id] = []

        OD[id].append(od)

OD_l = []
for id in OD:
    OD_l.append(OD[id])

slc = slice(0,None,3)

fig, ax = plt.subplots(figsize=(10,10))
plt.violinplot(OD_l[slc], positions=list(OD.keys())[slc], widths=0.01)
plt.title(f'connectivity = {p_con*100} %')
plt.show()
```
