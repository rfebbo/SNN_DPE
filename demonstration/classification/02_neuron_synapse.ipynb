```python
from snn_dpe.neuron_synapse import Neuron, Synapse
from snn_dpe.tools.plotting import plot_network

# create three neurons
n0 = Neuron(threshold=0.75, leak=0.01)
n1 = Neuron(threshold=0.75, leak=0.01)
n2 = Neuron(threshold=0.75, leak=0.01)

# connect n0 to n1 and n2 with two synapses
s0 = Synapse(n0, n1, weight=0.5, delay=0)
s1 = Synapse(n0, n2, weight=0.75, delay=1)
n0.add_synapse(s0)
n0.add_synapse(s1)

# put the neurons in a list
neurons = [n0, n1, n2]

# calls update on a list of neurons and prints if the neuron spiked and its membrane voltage
def update_and_print(neurons):
    for i in range(len(neurons)):
        print(f'n{i}: ', end='') 
        if neurons[i].update():
            print(f'spiked! ', end='')
        print(f'membrane voltage {neurons[i].membrane_voltage}')
    print()

print('update 1')
print('Running updates')
update_and_print(neurons)

print('update 2')
print('applying potential of 0.3 to n0 and running updates')
n0.apply_potential(0.3)
update_and_print(neurons)

print('update 3')
print('applying potential of 0.3 to n0 and running updates')
n0.apply_potential(0.3)
update_and_print(neurons)

print('update 4')
print('applying potential of 0.3 to n0 and running updates')
n0.apply_potential(0.3)
update_and_print(neurons)

print('update 5')
print('Running updates')
update_and_print(neurons)

print('update 6')
print('Running updates')
update_and_print(neurons)


plot_network(neurons)
```
