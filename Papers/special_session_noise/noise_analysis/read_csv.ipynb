```python
import csv
import numpy as np
import matplotlib.pyplot as plt

results = []

for i in range(10):
    with open(f'MG_noise_vs_MSE_dataset_{i}.csv', 'r') as f:
        reader = csv.reader(f, delimiter=',')
        tr_mses = []
        te_mses = []
        noise = []
        for row in reader:
            tr_mses.append(float(row[0]))
            te_mses.append(float(row[1]))
            noise.append(float(row[2]))

        results.append((tr_mses, te_mses, noise))

results = np.asarray(results)
(tr_mses, te_mses, noise_axis) = np.mean(results, axis=0)



plt.plot(noise_axis, tr_mses, label='Training')
plt.plot(noise_axis, te_mses, label='Testing')
plt.legend()
plt.xlabel('Noise std. dev.')
plt.ylabel('Mean Squared Error')
# plt.ylim(0, 100)
# plt.yscale('log')
plt.show()
```
